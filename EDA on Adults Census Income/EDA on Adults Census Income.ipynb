##### How does age,sex,marital status work together ?

Does income vary by sex within the same age and marital group?

Are married women earning less or more than single men in certain age groups?

Is age impacting marital status distributions differently for men and 
women?

```python
df['age_bin'] = pd.cut(df['age'],bins=[15,25,35,45,55,65,90],
                       labels=['15-25','25-35','35-45','45-55','55-65','65-90'])

grouped = df.groupby(['age_bin','sex','marital.status'])['income'].value_counts(normalize=True).unstack().fillna(0)
grouped = grouped.reset_index()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.barplot(data=grouped,x='marital.status',y='>50K',hue='sex')
plt.title("Proportion of >50K Income By Marital Status and Sex")
plt.xticks(rotation=45)
plt.ylabel('Proportion earning >50K')
plt.tight_layout()
plt.show()
```

```python
g = sns.catplot(data=grouped,x=df['marital.status',y='>50K',hue='sex',col='age_bin',
               kind='bar',col_wrap=3,height=4,aspect=1.2)

g.set_titles("{col_name}")
g.set_xticklabels(rotation=45)
g.fig.suptitle("Income >50K by Marital Status,Sex and Age Group",y=1.03)
plt.tight_layout()
plt.show()
```
