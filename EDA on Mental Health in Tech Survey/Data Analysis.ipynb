✅ What is Skewness?

Skewness is a measure of asymmetry in the distribution of a variable.

It tells us how the data is "tilted":

Are values bunched on one side?
Is there a long tail on one side?
🔍 Types of Skewness

Type	Description	Example
Symmetric	Mean ≈ Median ≈ Mode	Height of adults
Right-skewed	Tail extends to the right (positive)	Income, Age
Left-skewed	Tail extends to the left (negative)	Test scores (easy test)
🎯 Why is Skewness Important?

1. Affects Choice of Summary Statistics
In symmetric distributions, mean ≈ median
In skewed distributions:
Right-skew → mean > median
Left-skew → mean < median
→ Use median for better central tendency when skewed
2. Influences Data Transformation
Many ML models assume normality (e.g., linear regression, k-means).
If data is skewed, you might:
Apply log, sqrt, or Box-Cox transformations
Normalize or standardize features
3. Impacts Model Performance
Algorithms like Linear Regression, LDA, Naive Bayes work better with symmetrically distributed features.
Skewed input can lead to:
Biased coefficients
Poor generalization
4. Outlier Detection
High skewness can indicate presence of outliers.
Especially in right-skewed data: large, rare values (e.g., income = $1M)
5. Real-World Interpretation
Right-skewed Age → majority are young, few old
Right-skewed Income → few high earners dominate average
Left-skewed Test Score → most people scored very high (easy test)
✅ In short:

Skewness helps you understand distribution shape, choose appropriate statistics, detect anomalies, and improve model readiness.

 ## Measure Categorical Variability 

```python
from scipy.stats import entropy

# Convert counts to probabilities
probs = df['work_interfere'].value_counts(normalize=True,dropna=True)
entropy_val = entropy(probs)
print("Entropy (variability measure) : ",entropy_val)
```

Higher entropy ⇒ More variability (responses are evenly distributed)
Lower entropy ⇒ Less variability (one or two dominate)

🔍 How to Interpret Variability

Let’s say:

Sometimes     450
Often         250
Rarely        100
Never         80
NaN           120
"Sometimes" dominates → lower variability
If all values were ~equal → high variability
If a few people chose “Never” or “Rarely” → those are minority opinions → possible polarization
✅ Final Answer Template:

##### Clean up ambiguous gender entries (like "male ", "M", etc.)

```python
def clean_gender(g):
    g = str(g).strip().lower()
    if g in ['M', 'Male', 'male','m', 'Male-ish', 'maile','something kinda male?','Cis Male','Mal', 'Male (CIS)','male leaning androgynous',
       'Male ', 'Man','msle','cis male','Malr','Cis Man']:
        return "male"
    elif g in ['Female','female','Trans-female', 'Cis Female', 'F','f','queer/she/they','Femake', 'woman','Female ','cis-female/femme','Trans woman','Female (trans)','Female (cis)','femail']:
        return "female"
    else:
        return "others"
```

* What’s the correlation between family_history and seeking treatment?

To find the correlation between family_history and seeking treatment, you'll need to convert both columns from categorical ("Yes"/"No") into numeric (1/0), and then compute their Pearson correlation.

```python
#Map Yes/No to 1/0
df_corr = df[['family_history','treatment']].dropna()

df_corr['family_history_num'] = df_corr['family_history'].map({'Yes':1,'No':0})
df_corr['treatment_num'] = df_corr['treatment'].map({'Yes':1,'No':0})

#Compute Correlation
correlation = df_corr['family_history_num'].corr(df_corr['treatment_num'])
print(f"Correlation between family history and treatment: {correlation:.3f}")
```

🧠 Interpretation:
Correlation close to +1 → Strong positive relationship: people with family history are more likely to seek treatment.
Correlation near 0 → Little or no linear relationship.
Negative correlation (unlikely here) → Reverse trend.
